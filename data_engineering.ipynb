```python
#convert time_stamp to datetime in weather data
weather_df['time_stamp']=pd.to_datetime(cab_df['time_stamp'],unit='ms')
weather_df.head()
```

```python
#create new column for date
weather_df['date']=cab_df['time_stamp'].dt.strftime('%Y/%m/%d')
weather_df['occurence_time']=cab_df['time_stamp'].dt.strftime('%H:%M')
```

```python
weather_df.head(10000)
```

```python
#Group observations by date and take the average of observations within 24 hours
avgWeather=weather_df.groupby('date').agg(avg_temp=('temp', 'mean'), avg_rain=("rain",'mean'),avg_humidity=('humidity','mean'),observations=('date','count'))
avgWeather= avgWeather.reset_index()
avgWeather
```

```python
#filter weather dataset from Dec13 2018 - Dec 16 2018
start_date = '2018/12/13'
end_date = '2018/12/16'

filterr = (avgWeather['date'] >= start_date) & (avgWeather['date'] <= end_date)  
recent_weather = avgWeather.loc[filterr]
recent_weather= recent_weather.reset_index()
recent_weather=recent_weather.drop('index',axis=1)
recent_weather
```

```python
#filter rideshare dataset from Dec13 2018 - Dec 16 2018
start_date = '2018/12/13'
end_date = '2018/12/16'

mask = (cab_df['date'] > start_date) & (cab_df['date'] <= end_date)

cab_data = cab_df.loc[mask]
cab_data.reset_index(drop=True,inplace=True)
cab_data#134,784 rides between Dec13 & Dec 16
##add weather data for dec 13 14 15 16

#ride_df_groupby=df.groupby['date']
#ride_df
```

```python
#Remove time_stamp from cab data
cab_data=cab_data.drop(['time_stamp'],axis=1)
cab_data
```

```python
#Convert date column in cab_data to datetime
cab_data['date']=pd.to_datetime(cab_data['date'])
cab_data.dtypes
#convert date column in recent_weather data to datetime
#recent_weather['date']=pd.to_datetime(recent_weather['date'])
#df2['captureDate']  = pd.to_datetime(df2['captureDate'])
```

```python
try:
#convert date column in recent_weather data to datetime
    recent_weather['date']=pd.to_datetime(recent_weather['date'])
except KeyError:
    print('Ignoring key error for date....')
recent_weather.dtypes
```

```python
try:
#Join weather data with cab data
    cabweather=pd.merge(cab_data,recent_weather[['date','avg_temp','avg_rain','avg_humidity']],
                   on='date',how='inner')
except KeyError:
    print('Ignoring key error for date variable....')
    

cabweather.head(15)
```

```python
##Rearranging columns for merged dataframe
cabweather=cabweather[['id','product_id','date','cab_type','name','distance','price','surge_multiplier','source','destination','occurence_hour','avg_temp','avg_rain','avg_humidity']]

cabweather.head(10)
```

```python
#Save merged data as csv
cabweather.to_csv('Resources/cabweather.csv')
```

```python
#Save merged data as json file
cabweather.to_json('Resources/cabweather.json')
```

```python
###using scraped data from craigslist
###clean and format scraped data from craigslist
car_df.head()
car_df=car_df[['Title','Prices','Dates','Links']]
df_date = car_df["Dates"].str.split(expand=True).rename(columns={0:"Date", 1:"Time"})
df_date['Date'] = pd.to_datetime(df_date['Date'])
df_date.head()
```

```python
# create the new database and change the date type of price
car_listing = pd.DataFrame({"Date":df_date.Date, "Title":car_df.Title, "Price":(car_df.Prices), "Links":car_df.Links})

# convert price to number and drop abnormal date
car_listing['Price'] = car_listing['Price'].str.replace('$', '')
car_listing['Price'] = pd.to_numeric(car_listing['Price'])
car_listing = car_listing[car_listing['Price'] < 500000]
car_listing
```

```python
#Change year of date variable to 2018 to match other Datasets
car_listing['Date']=car_listing['Date'] - pd.DateOffset(years=1)
car_listing.head()
```

```python
#Rename column from 'Date' to 'date' to match other datasets
car_listing=car_listing.rename(columns={"Date":"date"})
car_listing.to_csv('Resources/carlisting.csv')
car_listing
```

```python
#Merge weather data with car_listing
try:
#Join weather data with cab data
    carweather=pd.merge(car_listing,recent_weather[['date','avg_temp','avg_rain','avg_humidity']],
                   on='date',how='inner')
except KeyError:
    print('Ignoring key error for date variable....')
    
```
