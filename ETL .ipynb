```python
import pandas as pd
import psycopg2
from sqlalchemy import create_engine
```

```python
#store csv into dataframe.
df_cab = pd.read_csv('Resources/cabweather.csv')
df_carlisting = pd.read_csv('Resources/carlisting_weather.csv')
df_carlisting_g = pd.read_csv('Resources/carlisting_weather_grouped.csv')
```

```python
df_cab=df_cab[['id','product_id','date','cab_type','name','distance','price','surge_multiplier','source','destination','occurence_hour','avg_temp','avg_rain','avg_humidity']]
```

```python
df_carlisting=df_carlisting[['date','Title','Price','Links','avg_temp','avg_rain','avg_humidity']]
df_carlisting.head()
```

```python
df_carlisting_g=df_carlisting_g[['date','num_of_listing','avg_price','avg_temp','avg_rain','avg_humidity']]
df_carlisting_g.head()
```

```python
connection_string = "postgres:toronto212@localhost:5432/ETLProject"
engine=create_engine(f'postgresql://{connection_string}')
connection = engine.connect()



#connection_string="postgres:toronto212@localhost:5432/county_db"
#engine=create_engine(f'postgresql://{connection_string}')
```

```python
df_cab.to_sql(name='cabweather', con=connection, if_exists='append', index=False)
```

```python
#load dataframe into database
df_carlisting.to_sql(name='carlisting_weather', con=connection, if_exists='append', index=False)
```

```python
df_carlisting_g.to_sql(name='carlisting_weather_grouped', con=connection, if_exists='append', index=False)
```
