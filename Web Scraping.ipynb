# Part 1 - Web Scraping to secure the heater sales data from Craigslist, Boston

```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
from splinter import Browser
import time 
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
titles = list()
prices = list()
dates = list()
links = list()
```

```python
link = 'https://boston.craigslist.org/search/sss?query=cars'
browser.visit(link)
```

```python
# Scrape the date info from the craigslist 

for x in range(24):
    
    html = browser.url
    response = requests.get(html)
    soup = bs(response.text, 'html.parser')

    results = soup.find_all('li', class_="result-row")

    for result in results:
        try:
        # Identify and return title of listing
            title = result.find('a', class_="result-title").text
        # Identify and return price of listing
            price = result.a.span.text
        # Identify and return link to listing
            link = result.a['href']
        # Identify and return the date
            date = result.time['datetime']

        # Print results only if title, price, and link are available
            if title and price and link and date:
            
                print('-------------')
                print(title)
                print(price)
                print(date)
                print(link)
            #Add the result to list     
                titles.append(title)
                prices.append(price)
                dates.append(date)
                links.append(link)
            
        except AttributeError:
            print("no item found")
        except:
            print("unknown error")
            
    time.sleep(5)
    browser.click_link_by_text('next > ')
```

```python
# convert the lists to Dataframe
car_df = pd.DataFrame({"Title":titles, "Prices":prices, "Dates":dates, "Links":links})
```

```python
# export to csv
car_df.to_csv('car_sales_by_date_raw_data.csv')
```

```python
# re-format the data and convert it to date type
import datetime
df_date = car_df["Dates"].str.split(expand=True).rename(columns={0:"Date", 1:"Time"})
df_date['Date'] = pd.to_datetime(df_date['Date'])
df_date.head()
```

```python
# create the new database and change the date type of price
car_listing = pd.DataFrame({"Date":df_date.Date, "Title":car_df.Title, "Price":(car_df.Prices), "Links":car_df.Links})

# convert price to number and drop abnormal date
car_listing['Price'] = car_listing['Price'].str.replace('$', '')
car_listing['Price'] = pd.to_numeric(car_listing['Price'])
car_listing = car_listing[car_listing['Price'] < 500000]
car_listing
```

```python
# count # of listing of heater by date
car_df_groupby =  car_listing.groupby('Date').agg(num_of_listing=('Title', 'count'), avg_price=("Price", 'mean'))
car_df_groupby['avg_price'] = car_df_groupby['avg_price'].round(0)
car_df_groupby = car_df_groupby.reset_index()

# offset 1 year to fit the other Dataset
car_df_groupby['Date_Acc'] = car_df_groupby['Date'] - pd.DateOffset(years=1)
```

```python
# finalize the dateset for the next part
car_df = car_df_groupby[['Date_Acc', 'num_of_listing', 'avg_price']]
car_df.to_csv('car_listing_from_Dec13_Dec16')
car_df
```
